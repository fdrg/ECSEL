# 3.  Country ranking   


```python
def make_pretty(styler):
    styler.set_caption("Main recipients")
    styler.format({'Amount':'â‚¬{:,.0f}','Country Share %':'{:,.2%}', 'Accum. %':'{:,.2%}', 'Contracts Share %':'{:,.0%}'})
    return styler
```

```python
aggregation = {
               'Projects':     ('Project',      pd.Series.nunique),
               'Coordination': ('Coordinator', 'sum'),
               'Amount':       ('Amount',      'sum')
              }
country_ranking = df.groupby(['Country'], as_index=True).agg(**aggregation).sort_values(['Amount'], ascending=False)

#country_ranking = df.groupby([df["Country"]]).sum()
#country_ranking = country_ranking.sort_values(['Amount'], ascending=False)[:12]

country_ranking['Country Share %']= country_ranking["Amount"]/country_ranking["Amount"].sum()
country_ranking['Accum. %']= country_ranking['Country Share %'].cumsum()
country_ranking['Contracts Share %']= country_ranking["Projects"]/number_of_contracts

country_ranking = country_ranking[:12]
```

```python
country_ranking.style.pipe(make_pretty)
```

# 4. Beneficiary Ranking   


```python
def make_pretty(styler):
    styler.set_caption("Main recipients, entities")
    styler.format({'Amount':'â‚¬{:,.0f}','Beneficiary Share %': '{:.2%}', 'Accum. %': '{:.2%}'}  )
    return styler
```

```python
aggregation = {
               'Contracts':    ('Project',      pd.Series.nunique),
               'Coordination': ('Coordinator', 'sum'),
               'Amount':       ('Amount',      'sum')
              }
beneficiary_ranking = df.groupby(['Country', 'Name'], as_index=True).agg(**aggregation).sort_values(['Amount'], ascending=False)
```

```python
def show_entity (entity, country, year):
    
    df_entity = df.loc[ (df['Beneficiary'].str.contains(entity)) & 
                        (df['Country'].str.contains(country)) & 
                        (df['Year'] == year) ]

    display(pd.pivot_table(df_entity, index = ['Subject'],
                           values = ['Amount','Coordinator'],
                           aggfunc = 'sum', fill_value = 'N/A',
                           margins = True, margins_name='Total'
                          ).sort_values(by='Amount', ascending=False)
           )
    
```
